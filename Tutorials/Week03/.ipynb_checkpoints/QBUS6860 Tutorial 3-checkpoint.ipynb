# Part 1: Monthly Temperature (Line Plot, Pie Chart, Distribution Plot)

Global Temperature Dataset contains the world historical temperature, available at https://datahub.io/core/global-temp. Your task is to choose the appropriate way to visualise the data.

## Step 1: Loading Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
sns.set(style="darkgrid")
```

## Step 2: Loading Data With Pandas

```python
df=pd.read_csv('monthly-temp.csv')
print(df)
```

You shall always have an overview for the loaded data. From the printout, we can see there are different types of mean temperatures. We will need to figure out how many are there and group the datapoints accordingly.

```python
print(df['Source'].unique())
```

 We see that we have 2 sources: one from GCAG and one from GISTEMP

## Step 3: Separating the data according to Source

```python
df1 = df[df['Source'] == 'GCAG']
df2 = df[df['Source'] == 'GISTEMP']
print(df1)
print(df2)
```

## Step 4: Draw Line Charts

```python
fig=px.line(df1,x='Date',y='Mean',title='Global Temperature (Source: GCAG)')
fig.show()
# If you are working on ED, please comment out the following two lines, 
# remove the above as ED does not support dynamic figures
#df1[::-1].plot(x='Date', y='Mean', title='Global Temperature (Source: GCAG)')
#plt.savefig("Result01.png")
```

```python
fig=px.line(df2,x='Date',y='Mean',title='Global Temperature (Source: GISTEMP)')
fig.show()

# If you are working on ED, please comment out the following two lines, 
# remove the above two as ED does not support dynamic figures
# df2[::-1].plot(x='Date', y='Mean', title='Global Temperature (Source: GISTEMP)') 
# plt.savefig("Result02.png")
```

## Step 5: Draw the Disctribution of Temperatures

```python
fig,axs=plt.subplots(1,2,figsize=(8,5))

sns.histplot(data=df1,x="Mean",color="red",label="GCAG Temp", kde=True, ax=axs[0])
sns.histplot(data=df2,x="Mean",color="blue",label="GISTEMP", kde=True, ax=axs[1])

plt.show()
#plt.savefig('Result04.png')
```

# Part 2: Tertiary Education by Country (Bar Chart, Lolipop Chart)

## Step 1: Load the Data

```python
data=pd.read_csv('education-by-country.csv')
data.head()
```

There are 17 different countries and the statistics on their tretiary education.
We will need to plot the bar chart showing the education %.

## Step 2: Plot the Bar Chart

```python
countries = data['country']
  
share_educated = data['t_education_percent']

plt.figure(figsize=(9,5))

plt.xlabel("Country")
plt.ylabel("Percentage of People with Tertiary Education")
plt.title("Percentage of People with Tertiary Education by Country in 2021")
  
    
plt.xticks(rotation=90)
      
plt.bar(countries, share_educated)
plt.show()
```

## Step 3: Plot the Lolipop Chart

```python
countries = data['country']
  
share_educated = data['t_education_percent']

ordered_df = data.sort_values(by='t_education_percent', ascending=False)
  
# Make the plot
plt.figure(figsize=(9,5))
plt.xlabel("Country")
plt.ylabel("Percentage of People with Tertiary Education")
plt.title("Percentage of People with Tertiary Education by Country in 2021")
  
   
plt.xticks(np.arange(17),data['country'], rotation=90)
plt.stem(ordered_df['t_education_percent'])
plt.show()
```
